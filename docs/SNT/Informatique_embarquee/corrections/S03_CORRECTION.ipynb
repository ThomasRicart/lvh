## ‚úÖ PROJET 1 ‚Äî üå°Ô∏è Alerte canicule
> *Affiche une alerte si la temp√©rature d√©passe un seuil ; bouton pour changer le seuil.*

### Tableau des composants attendu
| Composant | Nom | R√¥le |
|-----------|-----|------|
| Capteur | Thermom√®tre interne | Mesure la temp√©rature ambiante |
| IHM ‚Äî Entr√©e | Bouton A | Augmenter le seuil d'alerte |
| IHM ‚Äî Entr√©e | Bouton B | Diminuer le seuil d'alerte |
| IHM ‚Äî Sortie / Actionneur | Matrice LEDs | Affiche l'√©tat (normal/alerte) et le seuil |

### Algorithme attendu
```
Initialisation :
  seuil = 25

Boucle infinie :
  Lire la temp√©rature ‚Üí t
  Si bouton A appuy√© :
    seuil = seuil + 1
    Afficher le seuil pendant 1s
  Sinon si bouton B appuy√© :
    seuil = seuil - 1
    Afficher le seuil pendant 1s
  Sinon :
    Si t > seuil :
      Afficher Image.SKULL + d√©filer "ALERTE!"
    Sinon :
      Afficher Image.HAPPY
  Attendre 200 ms
```

### Code de r√©f√©rence
```python
from microbit import *

seuil = 25

while True:
    t = temperature()
    
    if button_a.was_pressed():
        seuil = seuil + 1
        display.scroll(str(seuil) + "C", wait=False)
    
    elif button_b.was_pressed():
        seuil = seuil - 1
        display.scroll(str(seuil) + "C", wait=False)
    
    else:
        if t > seuil:
            display.show(Image.SKULL)
            display.scroll("ALERTE " + str(t) + "C", wait=False)
        else:
            display.show(Image.HAPPY)
    
    sleep(200)
```

**Am√©liorations possibles :** Emp√™cher le seuil de descendre en dessous de 0¬∞C ou de d√©passer 50¬∞C.

## ‚úÖ PROJET 2 ‚Äî üéÆ Jeu de r√©action
> *Affiche une image al√©atoire ; bouton A = bonne r√©ponse, bouton B = mauvaise ; compte le score.*

### Tableau des composants attendu
| Composant | Nom | R√¥le |
|-----------|-----|------|
| IHM ‚Äî Entr√©e | Bouton A | R√©pondre ¬´ Oui / Bonne r√©ponse ¬ª |
| IHM ‚Äî Entr√©e | Bouton B | R√©pondre ¬´ Non / Mauvaise r√©ponse ¬ª |
| IHM ‚Äî Sortie / Actionneur | Matrice LEDs | Affiche le symbole √† reconna√Ætre et le score |

### Code de r√©f√©rence ‚Äî version ¬´ Plus ou Moins rapide ¬ª
```python
from microbit import *
import random

score = 0
images = [Image.HAPPY, Image.SAD, Image.ANGRY, Image.SURPRISED, Image.HEART]
# La ¬´ bonne r√©ponse ¬ª est d'appuyer sur A quand c'est HAPPY
cible = Image.HAPPY

while True:
    img_choisie = images[random.randint(0, len(images) - 1)]
    display.show(img_choisie)
    sleep(1500)  # L'√©l√®ve a 1,5s pour r√©pondre
    
    if button_a.was_pressed():
        if img_choisie == cible:
            score = score + 1
            display.show(Image.YES)
        else:
            score = max(0, score - 1)
            display.show(Image.NO)
        sleep(500)
    
    elif button_b.was_pressed():
        # Afficher le score actuel
        display.scroll("Score: " + str(score), wait=False)
    
    sleep(100)
```

## ‚úÖ PROJET 3 ‚Äî ‚è±Ô∏è Minuteur de sport
> *Compte √† rebours programmable ; secousse pour d√©marrer/arr√™ter.*

### Tableau des composants attendu
| Composant | Nom | R√¥le |
|-----------|-----|------|
| Capteur | Acc√©l√©rom√®tre (geste shake) | D√©marre ou arr√™te le minuteur |
| IHM ‚Äî Entr√©e | Bouton A | Augmenter la dur√©e (+5 secondes) |
| IHM ‚Äî Entr√©e | Bouton B | Diminuer la dur√©e (-5 secondes) |
| IHM ‚Äî Sortie / Actionneur | Matrice LEDs | Affiche le temps restant |

### Code de r√©f√©rence
```python
from microbit import *

duree = 30       # dur√©e en secondes
en_cours = False # le minuteur tourne-t-il ?
restant = duree

display.show(duree)

while True:
    if accelerometer.was_gesture('shake'):
        en_cours = not en_cours  # bascule d√©marrer/arr√™ter
        if en_cours:
            restant = duree
    
    if button_a.was_pressed() and not en_cours:
        duree = duree + 5
        restant = duree
        display.show(duree)
    
    if button_b.was_pressed() and not en_cours:
        duree = max(5, duree - 5)  # minimum 5 secondes
        restant = duree
        display.show(duree)
    
    if en_cours:
        display.show(restant)
        sleep(1000)
        restant = restant - 1
        if restant < 0:
            en_cours = False
            display.scroll("FINI!", wait=False)
    else:
        sleep(100)
```

**Note :** Ce projet est le plus complexe ‚Äî r√©serv√© aux √©l√®ves avanc√©s ou √† pr√©senter comme d√©fi bonus.

## ‚úÖ PROJET 4 ‚Äî üß≠ Boussole affich√©e
> *Indique la direction (N/S/E/O) avec une fl√®che.*

### Code de r√©f√©rence
```python
from microbit import *

# Calibrage n√©cessaire au d√©marrage
compass.calibrate()

while True:
    angle = compass.heading()  # 0 = Nord, 90 = Est, 180 = Sud, 270 = Ouest
    
    if button_a.was_pressed():
        # Afficher la valeur exacte en degr√©s
        display.scroll(str(angle), wait=False)
    
    else:
        # Afficher une fl√®che directionnelle
        if angle < 45 or angle >= 315:
            display.show(Image.ARROW_N)   # Nord
        elif angle < 135:
            display.show(Image.ARROW_E)   # Est
        elif angle < 225:
            display.show(Image.ARROW_S)   # Sud
        else:
            display.show(Image.ARROW_W)   # Ouest
    
    sleep(200)
```

## ‚úÖ PROJET 5 ‚Äî üí° D√©tecteur de mouvement
> *Utilise l'acc√©l√©rom√®tre pour d√©tecter si la carte est pos√©e ou boug√©e.*

### Code de r√©f√©rence
```python
from microbit import *

nb_alertes = 0
display.show(Image.ASLEEP)

while True:
    if accelerometer.was_gesture('shake'):
        nb_alertes = nb_alertes + 1
        display.show(Image.SURPRISED)
        display.scroll("MVT!", wait=False)
    
    elif accelerometer.was_gesture('face up'):
        display.show(Image.HAPPY)    # carte √† plat, face vers le haut
    
    elif accelerometer.was_gesture('face down'):
        display.show(Image.SAD)      # carte retourn√©e
    
    if button_a.was_pressed():
        display.scroll(str(nb_alertes) + " mvts", wait=False)
    
    if button_b.was_pressed():
        nb_alertes = 0
        display.show(Image.ASLEEP)
    
    sleep(100)
```

## ‚úÖ PROJET 6 ‚Äî üé≤ Mini-jeu de m√©moire
> *Affiche une s√©quence de symboles √† r√©p√©ter avec A et B.*

### Code de r√©f√©rence ‚Äî version simplifi√©e (s√©quence fixe)
```python
from microbit import *
import random

# A=0, B=1 dans la s√©quence
def generer_sequence(longueur):
    return [random.randint(0, 1) for _ in range(longueur)]

def montrer_sequence(seq):
    for s in seq:
        if s == 0:
            display.show("A")
        else:
            display.show("B")
        sleep(600)
        display.clear()
        sleep(200)

niveau = 2
sequence = generer_sequence(niveau)
saisie = []

display.scroll("GO!", wait=False)
sleep(800)
montrer_sequence(sequence)

while True:
    if button_a.was_pressed():
        saisie.append(0)
        display.show(Image.YES)
        sleep(200)
    
    if button_b.was_pressed():
        saisie.append(1)
        display.show(Image.YES)
        sleep(200)
    
    if len(saisie) == len(sequence):
        if saisie == sequence:
            display.show(Image.HAPPY)
            display.scroll("Super! Niv " + str(niveau + 1), wait=False)
            niveau = niveau + 1
        else:
            display.show(Image.SAD)
            display.scroll("Perdu!", wait=False)
            niveau = 2
        sleep(1500)
        sequence = generer_sequence(niveau)
        saisie = []
        montrer_sequence(sequence)
    
    sleep(50)
```

**Note :** Ce projet utilise des listes et des fonctions ‚Äî tr√®s bon exercice pour les √©l√®ves √† l'aise en Python.

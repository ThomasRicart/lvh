# 2. Présentation des dictionnaires

Un **dictionnaire** (`dict` en Python) est une structure de données non ordonnée (en termes d'indexation) qui stocke des couples **Clé : Valeur**.

Contrairement aux listes où l'on utilise des **indices** (0, 1, 2...), on utilise ici des **clés** (souvent des chaînes de caractères) pour retrouver une information.

**Synthaxe:**

* Le dictionnaire vide est `{}`
* le dictionnaire est de type `dict`
* chaque paire **clé, valeur** est spéarée par une virgule
* la **clé** et la **valeur** sont séparées par deux points `:`

```python
mon_dico = {}                           # dictionnaire vide
mon_dico = {'note': 19}                 # clé: 'note', valeur: 16
mon_dico = {'nom': 'Alice', 'note': 19} # deux couples clé: valeur
```

* pour accéder aux valeurs du dictionnaire, il suffit d'appeler la clé

```python
>>> mon_dico['note']
19
>>> mon_dico['nom']
'Alice'
```

* pour créer un couple **clé: valeur** il suffit de taper:

```python
mon_dico['age'] = 16
```

Le dictionnaire est maintenant modifié

```python
>> mon_dico
{'nom': 'Alice', 'note': 19, 'age': 16}
```

**Remarque:**
Bien que lorsqu'on observe un dictionnaire, on ait l'impression que les couples **clé: valeur** soient ordonnés, en fait ils ne le sont pas. C'est juste un affichage.

!!! info Exécutez les cellules suivantes et observez ce qu'il se passe


!!!

````python
mon_dico = {}
print(mon_dico)
print(type(mon_dico))
print(len(mon_dico))

### le dictionnaire vide est {}
### le dictionnaire est de type dict
### le dictionnaire contient 0 éléments
````

````python
mon_dico = {'nom': 'Alice', 'note': 19, 'age': 16}
print(mon_dico)
print(len(mon_dico))

### le dictionnaire n'est plus vide
### le dictionnaire contient 3 éléments
````

````python
mon_dico['classe'] = '1NSI'
print(mon_dico)
print(len(mon_dico))

### le dictionnaire contient 4 éléments
### On a ajouté le couple 'classe': '1NSI'
````

````python
mon_dico['note'] = 18
print(mon_dico)
print(len(mon_dico))

### le dictionnaire contient toujours 4 éléments
### On a modifié la valeur de la clé 'note'
````

# 3. Itération sur les dictionnaires

Puisque les dictionnaires ne sont plus ordonnés, on ne peut pas itérer sur les dictionnaires en utilisant les indices.

On va donc  utiliser les méthodes suivantes:
* `.keys()` pour accéder aux clés
* `.values()` pour accéder aux valeurs
* `.items()` pour accéder au couple **cle: valeur**

Les `keys()`, `values()` et `items()` sont des **itérables**.

!!! info Exécutez les cellules suivantes et observez ce qu'il se passe


!!!

````python
mon_dico = {'nom': 'Alice', 'note': 19, 'age': 16}

print(mon_dico.keys())                  # affiche les clés
print(mon_dico.values())                # affiche les valeurs
print(mon_dico.items())                 # affiche les couples cle: valeur
````

````python
# Itération sur les clé (mon_dico.keys())
# la variable cle prend toutes les valeurs de mon_dico.keys()
# On aurait pu donner n'importe quel nom aux clés
# affichage de chaque cle
# affichage de chaque valeur par mon_dico[cle]

for cle in mon_dico.keys():
    print(cle, mon_dico[cle])
````

````python
# Itération sur les valeurs (mon_dico.values())
# la variable val prend toutes les valeurs de mon_dico.values()
# On aurait pu donner n'importe quel nom aux valeurs
# affichage de chaque valeur

for val in mon_dico.values():
    print(val)
````

````python
# Itération sur les clés et valeurs (mon_dico.items())
# les variables cle et val prennent toutes les valeurs de mon_dico.items()
# On aurait pu donner n'importe quel nom aux clés et aux valeurs
# affichage de chaque clé et valeur

for cle, val in mon_dico.items():
    print(cle, val)
````

!!! note Itération sur un dictionnaire

Il existe trois façons d'itérer sur les dictionnaires:

```python
for cle in dico.keys():
    print(cle)
    
for val in dico.values():
    print(val)

for cle, val in dico.items():
    print(cle, val)

Comme `dico.items()` est un itérable constitué de `tuples`, `cle, val` prend les valeurs de chaque `tuple`.

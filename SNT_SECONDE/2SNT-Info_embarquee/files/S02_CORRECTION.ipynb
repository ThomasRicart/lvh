## ‚úÖ CORRECTION ‚Äî Activit√© 2 / Question 2 ‚Äî Analyse du code avec les boutons

```python
from microbit import *
while True:
    if button_a.is_pressed():
        display.show(Image.HAPPY)
    elif button_b.is_pressed():
        display.show(Image.SAD)
    else:
        display.show(Image.ASLEEP)
    sleep(100)
```

**1. Pourquoi une boucle `while True` ?**  
Un syst√®me embarqu√© doit fonctionner **en permanence**, sans s'arr√™ter. La boucle `while True` est infinie : le programme tourne ind√©finiment, v√©rifiant les boutons √† chaque tour. Sans cette boucle, le programme s'ex√©cuterait une seule fois puis s'arr√™terait ‚Äî la carte ne r√©agirait plus.

**2. Que se passe-t-il si on n'appuie sur aucun bouton ?**  
La condition `else` s'ex√©cute : la carte affiche `Image.ASLEEP` (visage endormi). C'est l'√©tat ¬´ par d√©faut ¬ª du syst√®me.

**3. Que se passe-t-il si on appuie sur A et B en m√™me temps ?**  
La condition `if button_a.is_pressed()` est √©valu√©e **en premier**. Si elle est vraie, le `elif` n'est pas test√©. Donc si A et B sont appuy√©s simultan√©ment, c'est l'image HAPPY qui s'affiche (A prioritaire). Pour g√©rer ce cas explicitement, il faudrait ajouter une condition `if button_a.is_pressed() and button_b.is_pressed()` **avant** les autres.

## ‚úÖ CORRECTION ‚Äî Activit√© 2 / Mini-d√©fi : Le compteur de points

```python
from microbit import *

compteur = 0
display.show(compteur)

while True:
    if button_a.was_pressed():
        compteur = compteur + 1
        display.show(compteur)
    
    elif button_b.was_pressed():
        compteur = 0
        display.show(compteur)
    
    sleep(100)
```

**Points cl√©s √† expliquer aux √©l√®ves :**

- On utilise `was_pressed()` (et non `is_pressed()`) pour d√©tecter **un appui unique** : `was_pressed()` renvoie `True` une seule fois par appui, m√™me si le bouton reste enfonc√©. Avec `is_pressed()`, le compteur s'incr√©menterait tr√®s vite tant que le bouton est maintenu.
- La variable `compteur` est initialis√©e **avant** la boucle : c'est l'initialisation du syst√®me.
- `display.show(compteur)` affiche un entier directement (pas besoin de `str()` contrairement √† `display.scroll()`).

**Limitation :** Au-del√† de 9, un seul chiffre ne suffit plus. `display.show(12)` affichera `1` puis `2` rapidement. Pour afficher correctement un nombre √† deux chiffres, il faut utiliser `display.scroll(str(compteur))` ‚Äî bonne discussion √† avoir avec les √©l√®ves.

## ‚úÖ CORRECTION ‚Äî Activit√© 3 / Question 3 ‚Äî Code complet de la station m√©t√©o

```python
from microbit import *

while True:
    t = temperature()  # ‚ö†Ô∏è Ne pas nommer la variable "temperature" 
                       # (conflit avec la fonction temperature())
    
    if button_a.was_pressed():
        # Afficher la temp√©rature en Celsius
        display.scroll(str(t) + "C")
    
    elif button_b.was_pressed():
        # Afficher en Fahrenheit
        fahrenheit = t * 9 // 5 + 32  # calcul de la conversion
        display.scroll(str(fahrenheit) + "F")
    
    else:
        # Afficher l'emoji selon la temp√©rature
        if t < 18:
            display.show(Image.SURPRISED)   # froid
        elif t <= 25:
            display.show(Image.HAPPY)       # confortable
        else:
            display.show(Image.ANGRY)       # chaud
    
    sleep(500)
```

**‚ö†Ô∏è Erreur pi√®ge tr√®s fr√©quente :** Le code de d√©part utilisait `temperature = temperature()`, ce qui cr√©e un conflit de nom : Python √©crase la fonction `temperature()` avec la variable du m√™me nom. √Ä la deuxi√®me ex√©cution de la boucle, appeler `temperature()` provoque une erreur. Il faut toujours nommer la variable diff√©remment : `t`, `temp`, `temp_c`...

**Variante avec `display.scroll(..., wait=False)` :** Pour ne pas bloquer la boucle lors du scroll :
```python
display.scroll(str(t) + "C", wait=False)
```

## ‚úÖ CORRECTION ‚Äî Question finale (Bilan de s√©ance)

**1. Qu'est-ce qu'une IHM ?**  
Une IHM (Interface Homme-Machine) est l'ensemble des composants qui permettent √† un √™tre humain d'interagir avec un syst√®me informatique. Elle comprend des **entr√©es** (ce que l'humain envoie au syst√®me : boutons, clavier, capteur tactile) et des **sorties** (ce que le syst√®me renvoie √† l'humain : √©cran, LEDs, son).

**2. Deux entr√©es et deux sorties utilis√©es :**
- Entr√©es : bouton A, bouton B (et bonus : acc√©l√©rom√®tre/geste shake)
- Sorties : matrice 5√ó5 LEDs (images), texte d√©filant (`scroll`)

**3. Pourquoi `while True` dans les syst√®mes embarqu√©s ?**  
Un objet embarqu√© doit r√©agir **en permanence** aux √©v√©nements du monde ext√©rieur (appui sur un bouton, variation de temp√©rature...). La boucle infinie permet de v√©rifier en continu l'√©tat des capteurs et des IHM. Sans elle, le programme s'arr√™terait apr√®s une seule v√©rification.

**4. Temps r√©el et micro:bit :**  
Le **temps r√©el** signifie que le syst√®me doit r√©pondre √† un √©v√©nement dans un d√©lai garanti et suffisamment court pour que le syst√®me fonctionne correctement. La micro:bit fonctionne globalement en temps r√©el : elle v√©rifie les boutons et capteurs toutes les ~100ms (gr√¢ce au `sleep(100)`), ce qui est suffisant pour des applications √©ducatives. En revanche, pour des syst√®mes critiques (frein ABS, pilote automatique d'avion), les contraintes de temps r√©el sont beaucoup plus strictes (r√©ponse en quelques microsecondes).
